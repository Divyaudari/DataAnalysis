```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("District_wise_No_of_Major_Industries_as_on_Mar_2021_3.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Sl.No.'] = range(1, len(df) + 1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.District.value_counts()
```

```python
sns.set_style("whitegrid")
# 1. Bar Plot - Top 10 Districts with Most Industries
top_10_districts = df[df["District"] != "Grand Total"].nlargest(10, "Total")
sns.barplot(x="Total", y="District", hue="District", data=top_10_districts, palette="Blues_r", legend=False)
plt.xlabel("Total Industries")
plt.ylabel("District")
plt.title("Top 10 Districts with Most Industries")
plt.show()
```

Top 10 Districts with Most Industries in karnataka
-----

```python
industry_sums = df.drop(columns=["District", "Total"]).sum().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=industry_sums.values, y=industry_sums.index, hue=industry_sums.index, palette="viridis", legend=False)
plt.xlabel("Total Count")
plt.ylabel("Industry Type")
plt.title("Comparison of Different Industry Types")
plt.show()
```

the "Other" industry category overwhelmingly dominates the total industry count, making it difficult to compare the remaining industry types. 
---------

```python

# 3. Pie Chart - Industry Distribution for a District
district_name = "Bangalore Urban"
district_data = df[df["District"] == district_name].drop(columns=["District", "Total"]).squeeze()
plt.figure(figsize=(8, 8))
plt.pie(district_data, labels=district_data.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title(f"Industry Distribution in {district_name}")
plt.show()
```

The Garment sector (10.3%) is a major contributor, while Automobile (0.9%) and Pharmaceuticals have minimal shares. 
--------

```python
# 4. Box Plot - Industry Count Distribution
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns=["District", "Total"]), palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Industry Type")
plt.ylabel("Count")
plt.title("Industry Count Distribution")
plt.show()
```

Most industries have low counts, except a few like Rice Mills and Garments, which have higher outliers. 
--------

```python

# 5. Histogram - Frequency Distribution of Total Industries
plt.figure(figsize=(10, 6))
sns.histplot(df["Total"], bins=10, kde=True, color="blue")
plt.xlabel("Total Industries per District")
plt.ylabel("Frequency")
plt.title("Distribution of Total Industries Across Districts")
plt.show()
```

This indicates industrial concentration in specific regions while many districts have minimal industrial presence
-----

```python
# 6. Scatter Plot - Steel vs Automobile Industries
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Steel"], y=df["Automobile"], hue=df["Total"], size=df["Total"], sizes=(20, 200), palette="coolwarm")
plt.xlabel("Steel Industries")
plt.ylabel("Automobile Industries")
plt.title("Correlation Between Steel and Automobile Industries")
plt.show()
```

Most points are clustered near the origin, indicating that many districts have minimal industry presence. 
-------

```python
# 7. Heatmap - Correlation Matrix
plt.figure(figsize=(10, 6))
corr_matrix = df.drop(columns=["District"]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Industries")
plt.show()
```

The heatmap represents the correlation matrix of different industries, showing how strongly they are related. Darker red shades indicate a higher positive correlation, while blue shades signify weaker relationships. Steel, automobile, pharmaceuticals, and cement industries exhibit strong interdependence, suggesting that growth in one sector likely influences others.
-------

```python
# 8. Stacked Bar Plot - Industry Composition in Top 5 Districts
top_5_districts = df.nlargest(5, "Total").set_index("District")
industries = top_5_districts.drop(columns=["Total"])
plt.figure(figsize=(12, 6))
industries.T.plot(kind="bar", stacked=True, colormap="tab10", figsize=(12, 6))
plt.xlabel("Industry Type")
plt.ylabel("Number of Industries")
plt.title("Industry Composition in Top 5 Districts")
plt.legend(title="District", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation=45)
plt.show()
```

indicating a large number of diverse industries. Bangalore Urban and Grand Total contribute significantly to this category. Other industry types, such as Automobiles, Steel, and Rice Mills, have relatively low counts, showing limited industrial distribution.
------

```python
# 9. Violin Plot - Distribution of Industry Counts
plt.figure(figsize=(12, 6))
sns.violinplot(data=df.drop(columns=["District", "Total"]))
plt.xticks(rotation=45)
plt.xlabel("Industry Type")
plt.ylabel("Distribution")
plt.title("Industry Count Distribution")
plt.show()
```

Most other industries have low and compact distributions, suggesting limited variability. Garments, Casting, and Rice Mills show slight spread, while others remain nearly uniform.
------

```python
# 11. Line Plot - Trend of Major Industries Across Districts
plt.figure(figsize=(12, 6))
df.set_index("District")[["Steel", "Automobile", "Cement"]].plot(marker="o")
plt.xlabel("District")
plt.ylabel("Number of Industries")
plt.title("Trend of Major Industries Across Districts")
plt.xticks(rotation=90)
plt.show()
```

Automobile industries exhibit the highest fluctuations, peaking significantly in "Grand Total" and "Bangalore Urban." Steel industries show moderate variance, while Cement maintains a relatively stable trend across districts with minimal variation.
------

```python
# 12. Density Plot - Industry Density Estimation
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Total"], fill=True, color="purple")
plt.xlabel("Total Industries")
plt.ylabel("Density")
plt.title("Industry Density Estimation")
plt.show()
```

```python
# 13. Bubble Chart - Industry Comparison
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Cement"], y=df["Steel"], size=df["Total"], sizes=(20, 300), hue=df["Total"], palette="coolwarm")
plt.xlabel("Cement Industries")
plt.ylabel("Steel Industries")
plt.title("Industry Comparison - Bubble Chart")
plt.show()


```

```python
# 14. Grouped Bar Plot - Comparison of Cement and Steel Industries
df.plot(x="District", y=["Cement", "Steel"], kind="bar", figsize=(12, 6), colormap="viridis")
plt.xlabel("District")
plt.ylabel("Number of Industries")
plt.title("Comparison of Cement and Steel Industries Across Districts")
plt.xticks(rotation=90)
plt.show()
```

```python
# 15. Doughnut Chart - Industry Distribution in a District
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(district_data, labels=district_data.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"), wedgeprops=dict(width=0.4))
plt.title(f"Doughnut Chart: Industry Distribution in {district_name}")
plt.show()
```
